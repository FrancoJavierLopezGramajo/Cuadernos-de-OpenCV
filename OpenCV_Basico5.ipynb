````python
import cv2

#Capturar video de una camara
video_capture = cv2.VideoCapture(0)

while(video_capture.isOpened()):
  ret, video_images = video_capture.read()
  if ret==True:
    cv2.imshow('video', video_images)
    if cv2.waitKey(1) & 0xFF == ord('s'):
      break
video_capture.release()
cv2.destroyAllWindows()
````

#[**BGR en OpenCV**](https://www.youtube.com/watch?v=25zlS7-kD0w&list=PLBg7GSvtrU2MOLWM0bGU1_FT3LsJPyY-5&index=6)

Para manipular los colores de las im√°genes y videos en OpenCV se dispone de las siguientes funciones.<br/>

Bk ->(B=0,G=0,R=0)<br/>

B  ->(B=255,G=0,R=0)<br/>

G  ->(B=0,G=255,R=0)<br/>

R  ->(B=0,G=0,R=255)<br/>

Wh ->(B=255,G=255,R=255)<br/>

La combinaci√≥n de estos **colores de RGB** genera todos los colores posibles, desde el negro absoluto Bk hasta el blanco absoluto Wh.<br/>
Para llegar a generar estas mezclas de colores OpenCV hace uso de una librer√≠a matem√°tica de Python llamada **Numpy**.<br/>
**Numpy** le sirve a OpenCV para la representaci√≥n de im√°genes como matrices, y para el procesamiento de im√°genes desde la modificaci√≥n de arrays de Numpy. Asimismo, con Nympy nos permite mantenernos dentro del ecosistema de Python.<br/>

Cuando obtenemos una imagen en **OpenCV**, esta ya viene en RGB, con sus valores determinados, y por ello es necesario modificar estos valores en arrays para manipular la colorimetria. Aqu√≠ es importante aclarar que cuando OpenCV toma una imagen y la representa en matrices de arrays esta va a estar en BGR (y no en RGB como suele ser en editores de im√°genes). En la matriz de OpenCV primero va a estar el color azul -Blue-, luego el verde -Green-, y al final el color rojo -Red-.


###  <img src="https://cdn.jsdelivr.net/gh/devicons/devicon/icons/pycharm/pycharm-original.svg" height="30" alt="Pycharm logo"  /> Desarrollo del script de Python (entorno virtual)

````python
#Modificacion de BGR en OpenCV
import cv2
import numpy as np

bgr=np.zeros((300,300,3),dtype=np.unit8)  # <- Solo esta linea es para color negro
bgr[:,:,:]=(255,0,0)                      # <- Esta linea para azul
bgr[:,:,:]=(0,255,0)                      # <- Esta linea para verde
bgr[:,:,:]=(0,0,255)                      # <- Esta linea para rojo

cv2.imshow('BGR', bgr)
cv2.waitKey(0)
cv2.destroyAllWindows()
````

#[**Umbralizaci√≥n o Thresholding en OpenCV**](https://www.youtube.com/watch?v=pPcuzYqPMhE&list=PLBg7GSvtrU2MOLWM0bGU1_FT3LsJPyY-5&index=7)

Con esto se busca separar el fondo de la imagen del objeto de interes. Es el modo de segmentaci√≥n m√°s b√°sico en **OpenCV**. Fondo diferenciable de objetos.<br/>

Para esto necesitamos una imagen en escala de grises y localizar su umbral.
Es recomendable utilizar im√°genes en cuyos objetos se diferencien bien del fondo de la imagen, y que el fondo de esta imagen, en lo posible, no presente ruido a la imagen. Para trabajar con umbralizaci√≥n se suele apoyarse en la librer√≠a de **Numpy**.<br/>

La umbralizaci√≥n trabaja revisando la condici√≥n de que cada pixel es mayor o no del valor de umbral otorgado (que puede ser entre **0** y **255**). Entonces al ejecutarse el c√≥digo, este decide si es menor que el valor de umbral otorgado y le adjudica un cero **0** al pixel, que es el color negro. En tanto si el pixel supera el valor de umbral se le adjudica un uno **1** o color blanco (aunque este √∫ltimo podr√° ser modificado, es decir si en vez de colocar 255 ponemos otro valor entonces el pixel reflejar√° esa escala). Y as√≠ se ir√° diferenciando pixel por pixel.<br/>

Tenemos distintos tipos de aplicaci√≥n de la umbralizaci√≥n:<br/><br/>

**THRESH_BINARY**

nos solicita la imagen, el valor del umbral, el valor de umbral positivo, la librer√≠a de OpenCV a utilizar.<br/><br/>


**BINARY - BINARY INV**

en este caso nos va a solicitar la imagen, el valor del umbral, un valor de umbral positivo que ser√° con el que reemplazar√° al pixel, y la librer√≠a de **OpenCV**.<br/>

 **Binary INV** es lo mismo que el anterior pero con el efecto invertido en la imagen.<br/><br/>

 **TRUNCATE - TRUNCATE INV**

 Lo que hace el tipo de umbralizaci√≥n de **Truncate** es que cuando se detecta que un pixel ha sobrepasado el valor de umbral otorgado entonces toma a dicho valor para esa zona de pixeles mayores al umbral. En tanto si no cumple con la condici√≥n, es decir los pixeles que no sobrepasan al valor del umbral, se mostrar√°n esos pixeles tal cual estaban y la imagen quedar√° tal como lo es originalmente para los pixeles que no sobrepasan al umbral.
 En este caso el tercer valor que antes se nos solicitaba como valor de umbral positivo ahora ser√° ignorado.<br/><br/>


**TO ZERO - TO ZERO INV**

Aqu√≠ tambi√©n el tercer elemento ser√° ignorado, y solamante se considerar√° la imgen y el valor del umbral.

Una de las cosas m√°s divertidas de OpenCV es la detecci√≥n de colores en la imagen.<br/>
Primero hay que saber sobre el espacio de color **HSV** que es el que vamos a utilizar para detectar los colores.<br/><br/>
***
üé®HSV significa <br/><br/>

**H**ue / Matiz

**S**aturation / Saturaci√≥n

**V**alue / Valor o brillo
<br/><br/>

__H__ de **0** a **179**, comienza en rojo pasa por todos los colores y termina en rojo.

__S__ de **0** a **255**, comienza en blanco y termina en un color rojo fuerte

__V__ de **0** a **255**, comienza en negro y termina en blanco
***
<br/><br/>

Todo el rango por el que transcurre Hue, yendo de rojo a rojo pasando por diferentes matices, es la **distribuci√≥n de los componentes de Hue**.<br/>
Esta distribuci√≥n es la que vamos a utilizar para poder detectar los colores.<br/><br/>


A continuaci√≥n describimos una serie de pasos para obtener el color:<br/>

‚úÖ __1¬∞ Paso:__
  Es obtener la imagen, ya sea por un archivo de imagen o video, o bien a trav√©s de la captura de video streming desde una c√°mara.<br/>

‚úÖ __2¬∞ Paso:__
 Transformar de BGR A HSV.<br/>

‚úÖ __3¬∞ Paso:__
 Determinar el rango. Es decir, darle al c√≥digo los rango de los colores que deseamos detectar. Por ejemplo, se puede detectar el componente rojo de H.<br/>

‚úÖ __4¬∞ Paso:__
  Visualizaci√≥n. Aqu√≠ vamos a obtener el resultado de la visualizaci√≥n de los colores.
  Si por ejemplo le pedimos que identifique los colores rojos, al momento de visualizar la imagen se vera como blanco las partes detectadas pero como negro las que no cumplen con la condici√≥n.
<br/><br/>
Hay que considerar dos factores que influyen en la resoluci√≥n como la iluminaci√≥n y el fondo.
Por ejemplo si cuando configuramos un color es con una iluminaci√≥n media, va a actuar muy bien con esa iluminaci√≥n pero no tan bien con luz muy alta o muy baja.<br/>

#[**Marcar Centro y Coordenadas de una Zona Coloreada en OpenCV**](https://www.youtube.com/watch?v=CFuyXw-pfPY&list=PLBg7GSvtrU2MOLWM0bGU1_FT3LsJPyY-5&index=9)

Lo que vamos a hacer ahora es marcar el centro de la zona coloreada y delimitada con contornos de color y a la vez describir sus coordenadas dentro de la imagen.<br/>

Para esto se toma un c√≥digo en OpenCV marcando una zona segun su color en HSV y luego se agregan algunas l√≠neas que indicar√°n el centro y coordenadas.<br/>
Para esto se utiliza el m√©todo `circle()` de **OpenCV**, el cual requiere los siguientes argumentos:
primero el frame y luego el `x` y `y` marcando el punto central, <br/>
luego el radio que en este caso es 7, <br/>
le damos luego el color que en este caso es verde, el -1 final parece ser por default.<br/><br/>

A continuaci√≥n se va a necesitar un texto para las coordenadas y entonces utilizamos el m√©todo `putText()` y poner mucha atenci√≥n en sus argumentos:
primero el frame<br/> y luego tenemos  '{},{}',<br/> format(x,y) esta secci√≥n quiere decir que como las coordenadas van a ir cambiando dichas coordenadas en format(x,y) ir√°n reemplazando a '{},{}'<br/> Se ver√° de forma din√°mica que a medida que se vaya cambiando de posici√≥n se actualice el mensaje de coordenadas.<br/> Lo que sigue es el tipo de fuente, el grosor, el color de la fuente.   

###  <img src="https://cdn.jsdelivr.net/gh/devicons/devicon/icons/pycharm/pycharm-original.svg" height="30" alt="Pycharm logo"  /> Desarrollo del script de Python (entorno virtual)

````python
#Marcar el centro de un objeto de color azul y sus coordendas en la imagen

import cv2
import numpy as np

cap = cv2.VideoCapture(0)

azulBajo = np.array([100,100,20], np.uint8)
azulAlto = np.array([125,255,255], np.uint8)

while True:

  ret, frame = cap.read()

  if ret == True:
    frameHSV = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)
    mask = cv2.inRange(frameHSV, azulBajo, azulAlto)
    _,contornos,_ = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APROX_SIMPLE)

    for c in contornos:                                                          #Este for loop es para precisar cada contorno
      area = cv2.contourArea(c)
      if area > 3000:
        M = cv2.moments(c)
        if (M['m00']== 0): M['m00'] = 1                                    #Al hacerce luego una division y no puede ser denominador 0, entonces se agregaen esta linea de codigo el M['m00']=1
        x = int(M['m10'] / M['m00'])
        y = int(M['m01'] / M['m00'])
        cv2.circle(frame, (x,y), 7, [0,255,0], -1)
        font = cv2.FONT_HERSHEY_SIMPLEX
        cv2.putText(frame, '{},{}', format(x,y), (x+10,y), font, 0.75, (0,255,0), 1, cv2.LINE_AA)     #Mensaje que indica las coordenadas y que se actualiza segun cambien las coordenas de unicacion del objeto
        nuevoContorno = cv2.convexHull(c)
        cv2.drawContours(frame, [nuevoContorno], 0, (255,0,0), 3)


    cv2.imshow('frame', frame)

    if cv2.waitKey(1) & 0xFF == ord('s'):
      break

cap.release()
cv2.destroyAllWindows()
````

## L√≠nea

```
cv2.line(

  Imagen donde se va a visualizar,

  Punto inicial,

  Punto final,

  Color(BGR),

  Grosor de la l√≠nea)
```

Hay que tener en cuenta que **Numpy** trata a las im√°genes como matrices de filas y columnas. Mientras que en **OpenCV** para poner la posici√≥n y el lugar para que se visualicen las figuras se utilizar√°n las coordenadas de `x`  e  `y`<br/>

#**Detecci√≥n de Objeto con Figuras y Texto**
De lo visto anteriormente ahora veremos como aplicarlo en forma conjunta, es decir que seg√∫n el color detectado (o seg√∫n una condici√≥n preestablecida) ser√° detectado, marcado y con un texto aclaratorio del tipo de color detectado para este caso.<br/>

En este caso vale prestar atenci√≥n en el m√©todo `boundigRect()` con el argumento del iterador **c**. Esto lo que nos va a permitir es obtener las coordenadas necesarias para poder encerrar el color que querramos detectar en un rect√°ngulo.<br/>

All√≠ mismo observamos a las variables `x`, `y`, `w`, `h` que corresponden a las `coordenadas x` y m√°s el ancho y el alto.<br/>

Luego se lleva una decisi√≥n de procesos diferentes, para cuando se detecta el color azul o si se detecta el color amarillo. En el c√≥digo que est√° a continuaci√≥n si el color detectado es el azul entonces **OpenCV** va a dibujar un rect√°ngulo con las coordenas que ya hemos visto. As√≠ tambi√©n en el mismo condicional se agregan dos l√≠neas que se forman en `X` dentro del rect√°ngulo y finalmente el texto.<br/>
Si se detecta el color amarillo, entonces se detecta el punto central del √°rea comprometida, se determina el radio y se dibuja el c√≠rculo y se agrega el correspondiente texto.

````python
#Deteccion de objetos segun su color y figura geometrica con OpenCV en un Jupyter Notebook

import cv2
import numpy as np
import matplotlib.pyplot as plt

def dibujar_contornos(mask, color, frame, font):
    contornos, _ = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    for c in contornos:
        area = cv2.contourArea(c)
        if area > 3000:
            x, y, w, h = cv2.boundingRect(c)
            #Condicionales de decisi√≥n para Azul (rectangulo) o Amarillo (circulo)
            if color == (255, 0, 0):  # Azul
                cv2.rectangle(frame, (x, y), (x + w, y + h), color, 3)
                cv2.line(frame, (x, y), (x + w, y + h), color, 3)
                cv2.line(frame, (x + w, y), (x, y + h), color, 3)
                cv2.putText(frame, 'Azul', (x - 10, y - 10), font, 0.75, color, 2, cv2.LINE_AA)
            elif color == (0, 255, 255):  #Amarillo
                M = cv2.moments(c)
                if M['m00'] == 0:
                    M['m00'] = 1
                xcentro = int(M['m10'] / M['m00'])
                ycentro = int(M['m01'] / M['m00'])
                radio = int(np.sqrt(w * h / np.pi))  #Aproximar radio a partir del √°rea del rectangulo
                cv2.circle(frame, (xcentro, ycentro), radio, color, 3)
                cv2.putText(frame, 'Amarillo', (x - 10, y - 10), font, 0.75, color, 2, cv2.LINE_AA)

#Capturar un solo frame de la camara
cap = cv2.VideoCapture(0)

ret, frame = cap.read()

if not ret:
    print("No se pudo acceder a la c√°mara.")
else:
    #Definir los rangos de color en HSV
    azulBajo = np.array([100, 100, 20], dtype=np.uint8)
    azulAlto = np.array([125, 255, 255], dtype=np.uint8)

    amarilloBajo = np.array([15, 100, 20], dtype=np.uint8)
    amarilloAlto = np.array([45, 255, 255], dtype=np.uint8)

    font = cv2.FONT_HERSHEY_SIMPLEX

    #Convertir el frame de BGR a HSV
    frameHSV = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)

    #Crear mascaras
    maskAzul = cv2.inRange(frameHSV, azulBajo, azulAlto)
    maskAmarillo = cv2.inRange(frameHSV, amarilloBajo, amarilloAlto)

    #Dibujar contornos en el frame original
    dibujar_contornos(maskAzul, (255, 0, 0), frame, font)
    dibujar_contornos(maskAmarillo, (0, 255, 255), frame, font)

    #Liberar la c√°mara
    cap.release()

    #Mostrar la imagen con las figuras detectadas
    plt.figure(figsize=(10, 8))
    plt.title('Detecci√≥n de objetos por color y figura geom√©trica')
    plt.imshow(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.show()
````

Un contorno es una sucesi√≥n de puntos que rodea cierta area de inter√©s, en **OpenCV** vamos a utilizar a estos para encerrar √°reas de im√°genes binarias, es decir a blanco y negro. Dicha √°rea se va a mostrar a blanco o **255**. **OpenCV** utiliza para esto el m√©todo `cv2.findContours()` que recibe como par√°metros una imagen que esta binarizada, a esta la podemos obtener a trav√©s de la detecci√≥n de colores marcando los rangos que determinan cierto color, pero tambi√©n se puede emplear la umbralizaci√≥n o la detecci√≥n de bordes. El segundo p√°rametro es mode, que es el modo de recuperaci√≥n de los contornos, que pueden ser de las siguientes opciones:<br/> `cv2.RETR_LIST`, <br/>`cv2.RETR_EXTERNAL`, <br/>`cv2.RETR_CCOMP`, <br/>`cv2.RETR_TREE`.<br/> El tercer par√°metro es el m√©todo de aproximaci√≥n del contorno, para especificar dicho argumento utilizamos `cv2.CHAIN_NONE()`, almacena todos los datos del contorno y es ideal para obtener el contorno en s√≠, as√≠ como el m√©todo `cv2.CHAIN_APPROX_SIMPLE()` con el que le estamos diciendo que comprima los segmentos verticales, horizontales y diagonales y deje solamente los puntos finales o v√©rtices, de este modo no se almacenan puntos redundantes como en `cv2.CHAIN_NONE()`. M√°s all√° de que hay otros m√©todos para la aproximaci√≥n de contornos el `cv2.CHAIN_APPROX_SIMPLE()` suele ser el m√°s eficiente. El cuarto y √∫ltimo pr√°metro es offset, que se trata del desplazamiento opcional mediante el cual se desplaza cada punto del contorno.<br/>

Para la sustracci√≥n de im√°genes vamos a usar los m√©todos `cv2.subtrac()` y `cv2.absdiff()`.<br/>

En ambos m√©todos nos piden dos im√°genes, las que se van a restar.

Tambi√©n se puede utilizar el m√©todo bitwise para ver ciertas regiones de nuestra imagen y el resto en negro.
En el ejemplo del c√≥digo se captura una imagen que puede ser dibujo, fotograf√≠a o video y se va a observar la imagen contenida dentro del c√≠rculo y el resto quedar√° en negro.<br/>
Para esto se agrega el m√©todo `cv2.bitwise_and()` y el primer y segundo argumento debe ser la misma imagen y como tercer argumento va a ser la m√°scara que estabamos creando en el c√≠rculo en blanco que m√°s adelante ser√° reemplazada por la imagen. A esta l√≠nea de c√≥digo se la va a guardar en la variable **imgMask**. Esto puede usarse por ejemplo para darle un estilo a la transmisi√≥n de video.<br/>


Del mismo modo se puede hacer al rev√©s, en ves de obtener un c√≠rculo por donde se vea la imagen se lo va a cambiar por un c√≠rculo que oculte una parte de la imagen. Para realizar esto se aplica el operador **not**, es decir `cv2.bitwise_not(mask)`. Puede ser √∫til para ocultar una parte del video o imagen.

###  <img src="https://cdn.jsdelivr.net/gh/devicons/devicon/icons/pycharm/pycharm-original.svg" height="30" alt="Pycharm logo"  /> Desarrollo del script de Python (entorno virtual)

````python
#Utilizando mascaras de video o imagen
#Puede ser un circulo que recorte la imagen cuadrada
#o bien con bitwise_not() una mascara que oculte una parte del video o imagen

import cv2
import numpy as np

captura = cv2.VideoCapture(0)
mask = np.zeros((480,640), dtype=np.uint8)
mask = cv2.circle(mask,(320,240), 125, (255), -1)

mask = cv2.bitwise_not(mask)       #Esta linea de codigo solamente se utilizara para crear mascaras que oculten parte del video
cv2.imshow('Mask', mask)                   #esta linea de codigo es solamente para mostrar la linea anterior, la mascara que oculta

while (capture.isOpened()):
  ret, frame = capture.read()

  if ret == True:

    imgMask = cv2.bitwise_and(frame, frame, mask=mask)    #Creacion de la mascara
    cv2.imshow('video', imgMask)

    if cv2.waitKey(1) & 0xFF == ord('s'):
      break
  else: break


capture.release()
cv2.destroyAllWindows()
````

##**Translaci√≥n**
En primer lugar veremos c√≥mo trasladar una imagen, es decir c√≥mo cambiarla de ubicaci√≥n de un lugar a otro. Para ello necesitamos obtener el alto y ancho de la imagen, esto se logra haciendo que la imgen sea como una fila y columna, por eso vamos a usar el m√©todo `shape()` con el n√∫mero **1** para la columna que ser√≠a el ancho de la imagen, y con el n√∫mero **0** que ser√≠a la fila o sea el alto de la imagen.<br/>

Ahora bien necesitamos de una matriz de traslaci√≥n de 2 x 3, en donde la primera fila que se referir√° a la columna y se encontrar√° el componente en `x`.
Y en la segunda fila estar√° el componente en `y`.<br/>

matriz_traslacion.png

Considerando la matriz antes vista, en el c√≥digo lo haremos dentro de la variable **tras** a la que a trav√©s de **Numpy** le damos el arreglo probando primero con un `eje x` de **10** y un `eje y` de **100**.<br/>

A continuaci√≥n utlizaremos el m√©todo `cv2.warpAffine()`  al que le daremos tres par√°metros, la imagen de entrada, la matriz de trasformaci√≥n y el tama√±o de la imagen de salida.<br/><br/>

```
cv2.warpAffine(
  imagen de entrada,
  
  matriz de trasformaci√≥n,

  tama√±o de la imagen de salida)
```  

<br/><br/>
Este m√©todo aplica para las transformaciones de una imagen.<br/>
En nuestro caso del c√≥digo de abajo, la imagen de entrada ser√° **imagen10**, la matriz de transformaci√≥n ser√° la variable **tras** y la imagen de salida ser√° igual a la imagen original, todo esto ser√° alojado en la variable **image_out**
Ya solamente queda mostrarlo por pantalla.<br/>

Tambi√©n se puede trabajar con valores negativos tanto para `x` como para `y`

###  <img src="https://cdn.jsdelivr.net/gh/devicons/devicon/icons/pycharm/pycharm-original.svg" height="30" alt="Pycharm logo"  /> Desarrollo del script de Python (entorno virtual)

````python
#Traslacion de una imagen
import cv2
import numpy as np

imagen10 = cv2.imread('picachu.jpg')
ancho = imagen10.shape[1]  #Columna
alto = imagen10.shape[0]  #Fila

#Traslacion
tras = np.float32([[1,0,10],[0,1,100]])   #desplazamiento
image_out = cv2.warpAffine(imagen10, tras, (ancho, alto))

cv2.imshow("Imagen de entrada: ", imagen10)
cv2.imshow("Imagen de salida: ", image_out)

cv2.waitKey(0)
cv2.destroyAllWindows()
````

Como hab√≠amos dicho, el m√©todo `warpAffine()` sirve para transformaciones de la imagen, y por lo tanto tambi√©n se puede modificar el tama√±o de la misma. Para especificar el tama√±o de la imagen de salida podemos restar o sumar un valor al tercer par√°metro de `warpAffine()` sea el alto o el ancho, sea uno solo o a ambos.

###  <img src="https://cdn.jsdelivr.net/gh/devicons/devicon/icons/pycharm/pycharm-original.svg" height="30" alt="Pycharm logo"  /> Desarrollo del script de Python (entorno virtual)

````python
#Cambiar el tama√±o de una imagen
import cv2
import numpy as np

imagen11 = cv2.imread('picachu2.jpg')
ancho = imagen11.shape[1]  #Columna
alto = imagen11.shape[0]  #Fila

#Mover y cambiar el tama√±o a la vez
tras_dimensiones = np.float32([[1,0,10],[0,1,100]])   #desplazamiento
image_out = cv2.warpAffine(imagen11, tras_dimensiones, (ancho +25, alto +200))    #se cambia el tama√±o

cv2.imshow("Imagen de entrada: ", imagen11)
cv2.imshow("Imagen de salida: ", image_out)

cv2.waitKey(0)
cv2.destroyAllWindows()
````

##**Rotaci√≥n**
Vamos ahora a probar rotar la imagen. Al igual que en la traslaci√≥n necesitamos una matriz pero en este caso de rotaci√≥n. Para contruirla necesitaremos del m√©todo `cv2.gerRotationMatrix2D()` que calcula una matriz af√≠n de rotaci√≥n 2D. Adem√°s permite ajustar el centro de rotaci√≥n, as√≠ como la escala de la imagen. Los par√°mentros que usaremos para este m√©todo son: el centro de rotaci√≥n de la imagen de entrada, el √°ngulo de rotaci√≥n en grados (rotaci√≥n en sentido horario) y la escala o sea el valor de la escala isotr√≥pica.<br/>

```
gerRotationMatrix2D(

 Centro de rotacion de la imagen de entrada,

 √Ångulo de rotac√≠on en grados (rotaci√≥n en sentido horario),

 Escala)
```

Como se ve en el c√≥digo de abajo, en los argumentos del m√©todo `getRotationMatrix2D()` dividimos el ancho y el alto sobre 2. A la rotaci√≥n le dimos 15 grados, y a la escala 1, que quiere decir que se mantendr√° del mismo tama√±o de la imagen de entrada.<br/>

Una vez que ya tenemos la variable lista donde se aloja el resultado del m√©todo `getRotationMatrix2D()`, a continuaci√≥n usamos nuevamente el m√©todo `warpAffine()` para especificar c√≥mo se ver√° la imagen de salida.<br/>

Vale aclarar que de usarse n√∫meros negativos la imagen girar√° en el sentido contrario, es decir en sentido antihorario.

````python
#Rotar o girar una imagen
import cv2
import numpy as np

imagen12 = cv2.imread('picachu3.jpg')
ancho = imagen12.shape[1]  #Columna
alto = imagen12.shape[0]  #Fila

#Rotacion
girar = cv2.getRotationMatrix2D((ancho//2, alto//2),15,1)
image_out = cv2.warpAffine(imagen12, girar, (ancho, alto))

cv2.imshow("Imagen de entrada: ", imagen12)
cv2.imshow("Imagen de salida: ", image_out)

cv2.waitKey(0)
cv2.destroyAllWindows()
````

```
cv2.resize(

 Imagen de entrada,

 (nuevo ancho, nuevo alto) de la imagen,

 M√©todo de interpolaci√≥n)
```

##**Recorte**
Supongamos que queremos recortar la imagen de Picachu. Para ayudarnos con esta tarea vamos a imprimir por pantalla la informaci√≥n de filas y columnas que posee la imagen. Para esto nos ayudaremos de `shape()`<br/>
Obtendremos entonces tres valores por consola que describen la cantidad de filas, columnas y los canales de BGR.<br/>
En el caso de nuestro c√≥digo nos dio 360 filas, 640 columnas y 3 que corresponden a los tres canales BGR, o sea (360,640,3)<br/>

Dada esta informaci√≥n podemos emprezar a recortar. Hay que tener en cuenta que esta vez ya no estamos usando coordenadas de `x`  `y` para manipular la imagen, sino que vamos a tratar la imagen como una matriz, entonces lo platearemos de la siguiente manera: de las filas vamos a tomar de la fila inicial hasta la fila final, y de las columnas de la imagen tomaremos desde la columna inicial hasta la columna final.<br/>

    imagen[ fila inicial:fila final , columna inicial:columna final ]

Modificando estos valores de filas y columnas se ir√°n reflejando en la imagen el correspondiente recorte.

````python
#Trasladar, Rotar, Escalar y Recortar una Imagen con OpenCV en un Jupyter Notebook
import cv2
import numpy as np
import matplotlib.pyplot as plt

#Traslacion de una imagen
imagen10 = cv2.imread('picachu.jpg')

if imagen10 is None:
    print("Error: No se pudo cargar la imagen 'picachu.jpg'. Verifica la ruta del archivo.")
else:
    ancho = imagen10.shape[1]                                #columna
    alto = imagen10.shape[0]                                 #fila

    #Traslacion
    tras = np.float32([[1, 0, 10], [0, 1, 100]])             #desplazamiento
    image_out = cv2.warpAffine(imagen10, tras, (ancho, alto))

    #Mostrar imagenes
    fig, axes = plt.subplots(1, 2, figsize=(12, 6))

    axes[0].imshow(cv2.cvtColor(imagen10, cv2.COLOR_BGR2RGB))
    axes[0].set_title('Imagen Original')
    axes[0].axis('off')

    axes[1].imshow(cv2.cvtColor(image_out, cv2.COLOR_BGR2RGB))
    axes[1].set_title('Imagen Trasladada')
    axes[1].axis('off')

    plt.tight_layout()
    plt.show()

#------------------------------
#Cambiar el tama√±o de una imagen
imagen11 = cv2.imread('picachu2.jpg')

if imagen11 is None:
    print("Error: No se pudo cargar la imagen 'picachu2.jpg'. Verifica la ruta del archivo.")
else:
    ancho = imagen11.shape[1]                                #columna
    alto = imagen11.shape[0]                                 #fila

    #Mover y cambiar el tama√±o a la vez
    tras_dimensiones = np.float32([[1, 0, 10], [0, 1, 100]])  #desplazamiento
    image_out = cv2.warpAffine(imagen11, tras_dimensiones, (ancho + 25, alto + 200))  #se cambia el tama√±o

    #Mostrar imagenes
    fig, axes = plt.subplots(1, 2, figsize=(12, 6))

    axes[0].imshow(cv2.cvtColor(imagen11, cv2.COLOR_BGR2RGB))
    axes[0].set_title('Imagen Original')
    axes[0].axis('off')

    axes[1].imshow(cv2.cvtColor(image_out, cv2.COLOR_BGR2RGB))
    axes[1].set_title('Imagen Trasladada y Redimensionada')
    axes[1].axis('off')

    plt.tight_layout()
    plt.show()

#------------------------------
#Rotar o girar una imagen
imagen12 = cv2.imread('picachu3.jpg')

if imagen12 is None:
    print("Error: No se pudo cargar la imagen 'picachu3.jpg'. Verifica la ruta del archivo.")
else:
    ancho = imagen12.shape[1]                               #columna
    alto = imagen12.shape[0]                                #fila

    #Rotacion
    girar = cv2.getRotationMatrix2D((ancho // 2, alto // 2), 15, 1)
    image_out = cv2.warpAffine(imagen12, girar, (ancho, alto))

    #Mostrar imagenes
    fig, axes = plt.subplots(1, 2, figsize=(12, 6))

    axes[0].imshow(cv2.cvtColor(imagen12, cv2.COLOR_BGR2RGB))
    axes[0].set_title('Imagen Original')
    axes[0].axis('off')

    axes[1].imshow(cv2.cvtColor(image_out, cv2.COLOR_BGR2RGB))
    axes[1].set_title('Imagen Rotada (15 grados)')
    axes[1].axis('off')

    plt.tight_layout()
    plt.show()

#------------------------------
#Escalar una imagen
imagen13 = cv2.imread('picachu4.jpg')
imagen14 = cv2.imread('picachu5.jpg')

if imagen13 is None:
    print("Error: No se pudo cargar la imagen 'picachu4.jpg'. Verifica la ruta del archivo.")
elif imagen14 is None:
    print("Error: No se pudo cargar la imagen 'picachu5.jpg'. Verifica la ruta del archivo.")
else:
    #Escalado
    image_out = cv2.resize(imagen13, (800, 400), interpolation=cv2.INTER_CUBIC)

    #Escalado con redimension autom√°tica del ratio (usando solo OpenCV)
    #Calcular nuevo tama√±o manteniendo la proporcion
    width = 300
    height = int(imagen14.shape[0] * width / imagen14.shape[1])
    image_out2 = cv2.resize(imagen14, (width, height), interpolation=cv2.INTER_CUBIC)

    #Mostrar imagenes
    fig, axes = plt.subplots(2, 2, figsize=(12, 10))

    axes[0, 0].imshow(cv2.cvtColor(imagen13, cv2.COLOR_BGR2RGB))
    axes[0, 0].set_title('Imagen Original (para escalado fijo)')
    axes[0, 0].axis('off')

    axes[0, 1].imshow(cv2.cvtColor(image_out, cv2.COLOR_BGR2RGB))
    axes[0, 1].set_title('Imagen Escalada (800x400)')
    axes[0, 1].axis('off')

    axes[1, 0].imshow(cv2.cvtColor(imagen14, cv2.COLOR_BGR2RGB))
    axes[1, 0].set_title('Imagen Original (para escalado proporcional)')
    axes[1, 0].axis('off')

    axes[1, 1].imshow(cv2.cvtColor(image_out2, cv2.COLOR_BGR2RGB))
    axes[1, 1].set_title('Imagen Escalada (ancho=300, proporcional)')
    axes[1, 1].axis('off')

    plt.tight_layout()
    plt.show()

#------------------------------
#Recortar una imagen
imagen15 = cv2.imread('picachu6.jpg')

if imagen15 is None:
    print("Error: No se pudo cargar la imagen 'picachu6.jpg'. Verifica la ruta del archivo.")
else:
    print("Imagen shape = ", imagen15.shape)  #conversion a filas y columnas

    #Recorte
    image_out = imagen15[150:360, 200:400]  #[fila_i:fila_f, columna_i:columna_f]

    #Mostrar imagenes
    fig, axes = plt.subplots(1, 2, figsize=(12, 6))

    axes[0].imshow(cv2.cvtColor(imagen15, cv2.COLOR_BGR2RGB))
    axes[0].set_title('Imagen Original')
    axes[0].axis('off')

    axes[1].imshow(cv2.cvtColor(image_out, cv2.COLOR_BGR2RGB))
    axes[1].set_title('Imagen Recortada')
    axes[1].axis('off')

    plt.tight_layout()
    plt.show()
````

Para realizar las operaciones de modificaci√≥n de la imagen en un **Jupyter**, se reemplaz√≥ a imutils por una l√≥gica de **OpenCV** para mantener proporciones. Es entonces necesario hacer una escala proporcional sin imutils, para esto se calcula manualmente el nuevo tama√±o manteniendo la proporci√≥n de la imagen.
<br/>
`cv2.cvtColor()` se usa para convertir de BGR a RGB cuando se muestra una imagen en color y si esta est√° en BGR.<br/>
Tal como se hizo en c√≥digos para **Jupyter** anteriores, se a√±adi√≥ una verificaci√≥n para asegurar que las im√°genes se hayan cargado correctamente.<br/>
La visualizaci√≥n se hace con **Matplotlib**, se usan subplots para mostrar las im√°genes originales y los resultados de las operaciones.
